```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# Read the data
data = pd.read_csv('tmp/keywords.csv')
```

```python
# 将 DataFrame 转换为字典
data_dict = data.set_index('Keyword')['Frequency'].to_dict()

print(data_dict)
```

```python
font_path = 'PingFangSC-Regular.ttf'

wordcloud = WordCloud(width=1600, height=800, background_color='white', font_path=font_path, min_font_size=8).generate_from_frequencies(data_dict)

plt.figure(figsize=(20, 12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
# plt.savefig(f'{usee}.png')
```
